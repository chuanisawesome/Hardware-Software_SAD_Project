```python
import numpy as np
import pandas as pd
import csv
```

```python
# import csv data
f = open('data.csv', 'r')
df = pd.read_csv(f, header=None)
df[1]
```

```python
# separate column 2 by comma, and add it to the dataframe
# df.insert(1, 'data', df[1].str.split(', ').str[0])
df1 = df[1].str.split(', ', expand=True)
df1
```

```python

# remoec first character from each element in the first column
df1[0] = df1[0].str.slice(1)
# reomve the last character from each element in the last column
df1[4] = df1[4].str.slice(0, -1)

df1
```

```python
# convert df1 to numeric from string
df1 = df1.apply(pd.to_numeric)
```

```python
# plot graph for df1
import matplotlib.pyplot as plt
```

```python
# plot graph for df1
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots( nrows=5, ncols=1, sharex=True )
fig.set_size_inches(25, 15)
# plot exponential average
ax1.plot(df1[0], '-b')
ax2.plot(df1[1], '-b')
ax3.plot(df1[2], '-b')
ax4.plot(df1[3], '-b')
ax5.plot(df1[4], '-b')
```
